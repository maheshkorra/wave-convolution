In this question, we have been asked to plot the basic signals that were discussed in the class, through means of a program.

The basic signals that we identified were the following:-
1. The discrete delta signal (δ[n]);
2. The unit step signal (u[n]);
3. The real exponential signal (a^n, a being real).
4. (a) The real discrete sinusoidal signal (Asin(ωn + ø));
   (b) The complex signal e^(σ + j*ωn).

The following code is for basic signal 1 (δ[n]):-

```python
import matplotlib.pyplot as plt
import numpy as np

# Self - explanatory.
def delta():
    
    bit = [0]
    sig_val = [1 if index in bit else 0 for index in range(-10,11)]
    sig_indices = [index for index in range(-10,11)]
    
    plt.stem(sig_indices,sig_val)
    plt.xlabel('n')
    plt.ylabel('delta(n)')
    plt.title('Delta Signal')
    plt.show()
    
    
    
delta()
```

The following code is for basic signal 2 (u[n]):-

```python
import matplotlib.pyplot as plt
import numpy as np

# Self - explanatory.
def unit_step():
    
    sig_val = [1 if index > 0 else 0 for index in range(-10,11)]
    sig_indices = [index for index in range(-10,11)]
    
    plt.stem(sig_indices,sig_val)
    plt.xlabel('n')
    plt.ylabel('unit_step(n)')
    plt.title('Unit Step Signal')
    plt.show()
    
unit_step()
```

The following code is for basic signal 3 (a^n, a being real):-

```python
import matplotlib.pyplot as plt
import numpy as np

# Self - explanatory.
def realExponential(a):

    sig_indices = [index for index in range(-10, 11)]
    sig_val = [a**index for index in sig_indices]
    
    plt.stem(sig_indices, sig_val)
    plt.xlabel('n')
    #plt.ylabel('a^n, a = {}'.format(a)) -> this can be done as well
    plt.ylabel(str(a) + '^n')
    plt.title('Real Exponential Signal')
    plt.show()
    
realExponential(2)
```

The following code is for basic signal 4.(a) (Asin(ωn + ø):-

```python
import matplotlib.pyplot as plt
import numpy as np

# Self - explanatory.
def realDiscreteSinusoidal(magnitude, periodicity, phaseDifference):
    
    w_0 = (2*np.pi)/periodicity
    
    radPhaseDifference = phaseDifference * np.pi/180
    
    sig_indices = [index for index in range(-10, 11)]
    sig_val = [magnitude * np.sin(w_0 * index + radPhaseDifference) for index in sig_indices]
    
    plt.stem(sig_indices, sig_val)
    plt.xlabel('n')
    plt.ylabel('Xr[n]')
    plt.title('Real Discrete Sinusoidal')
    plt.show()
    
realDiscreteSinusoidal(magnitude = 10, periodicity = 10, phaseDifference = 45)
```

The following code is for basic signal 4.(b) (e^(σ + j*ωn)):-

```python
import matplotlib.pyplot as plt
import numpy as np

# Self - explanatory.
def complexDiscreteSinusoidal(magnitude, periodicity, phaseDifference):
    
    w_0 = (2*np.pi)/periodicity
    
    radPhaseDifference = phaseDifference * np.pi/180
    
    sig_indices = [index for index in range(-50, 51)]
    sig_val = [magnitude * np.cos(w_0 * index + radPhaseDifference) + magnitude * np.sin(w_0 * index + radPhaseDifference)*1j for index in sig_indices]
    
    
    sig_x = [comp.real for comp in sig_val]
    sig_y = [comp.imag for comp in sig_val]
    
    
    fig = plt.figure()
    axes = fig.gca(projection = '3d')
    axes.plot(sig_indices, sig_x, sig_y, label = 'Complex signal')
    axes.set_xlabel('n')
    axes.set_ylabel('Real')
    axes.set_zlabel('Imaginary')
    axes.legend()
    plt.show()
    
complexDiscreteSinusoidal(magnitude = 10, periodicity = 10, phaseDifference = 45)
```

```python
# Self - explanatory.

import matplotlib.pyplot as plt
import numpy as np


def delta():
    
    bit = [0]
    sig_val = [1 if index in bit else 0 for index in range(-10,11)]
    sig_indices = [index for index in range(-10,11)]
    
    plt.stem(sig_indices,sig_val)
    plt.xlabel('n')
    plt.ylabel('delta(n)')
    plt.title('Delta Signal')
    plt.show()

def unit_step():
    
    sig_val = [1 if index > 0 else 0 for index in range(-10,11)]
    sig_indices = [index for index in range(-10,11)]
    
    plt.stem(sig_indices,sig_val)
    plt.xlabel('n')
    plt.ylabel('unit_step(n)')
    plt.title('Unit Step Signal')
    plt.show()

def realExponential(a):

    sig_indices = [index for index in range(-10, 11)]
    sig_val = [a**index for index in sig_indices]
    
    plt.stem(sig_indices, sig_val)
    plt.xlabel('n')
    #plt.ylabel('a^n, a = {}'.format(a)) -> this can be done as well
    plt.ylabel(str(a) + '^n')
    plt.title('Real Exponential Signal')
    plt.show()
    
def realDiscreteSinusoidal(magnitude, periodicity, phaseDifference):
    
    w_0 = (2*np.pi)/periodicity
    
    radPhaseDifference = phaseDifference * np.pi/180
    
    sig_indices = [index for index in range(-10, 11)]
    sig_val = [magnitude * np.sin(w_0 * index + radPhaseDifference) for index in sig_indices]
    
    plt.stem(sig_indices, sig_val)
    plt.xlabel('n')
    plt.ylabel('Xr[n]')
    plt.title('Real Discrete Sinusoidal')
    plt.show()
    
    
def complexDiscreteSinusoidal(magnitude, periodicity, phaseDifference):
    
    w_0 = (2*np.pi)/periodicity
    
    radPhaseDifference = phaseDifference * np.pi/180
    
    sig_indices = [index for index in range(-50, 51)]
    sig_val = [magnitude * np.cos(w_0 * index + radPhaseDifference) + magnitude * np.sin(w_0 * index + radPhaseDifference)*1j for index in sig_indices]
    
    
    sig_x = [comp.real for comp in sig_val]
    sig_y = [comp.imag for comp in sig_val]
    
    
    fig = plt.figure()
    axes = fig.gca(projection = '3d')
    axes.plot(sig_indices, sig_x, sig_y, label = 'Complex signal')
    axes.set_xlabel('n')
    axes.set_ylabel('Real')
    axes.set_zlabel('Imaginary')
    axes.legend()
    plt.show()

delta()  
unit_step()  
realExponential(2)          
realDiscreteSinusoidal(magnitude = 10, periodicity = 10, phaseDifference = 45)
complexDiscreteSinusoidal(magnitude = 10, periodicity = 10, phaseDifference = 45)

    
```
